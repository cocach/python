```python
import urllib2
```

```python
response = urllib2.urlopen("https://zerostwo.github.io") #urlopen(url, data, timeout)
print response.read() #read()为response的一个方法
```

```python
# 上面的代码等价于下面
request = urllib2.Request("https://zerostwo.github.io")
response = urllib2.urlopen(request)
print response.read()
```

```python
import urllib
import urllib2
 
values = {
    "user":"201702420",
    "pwd":"981211",
    "lb":"S"
}
data = urllib.urlencode(values) 
url = "http://jiaowu.sicau.edu.cn/web/web/web/index.asp"
geturl = url + "?"+data
request = urllib2.Request(geturl)
response = urllib2.urlopen(request)
print response.read()
```
